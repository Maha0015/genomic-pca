```python
import subprocess

for i in range(1, 23):
    # Backup the original .bim file
    backup_command = f"cp Chr{i}_ma_filtered.bim Chr{i}_ma_filtered.bim.backup"
    subprocess.run(backup_command, shell=True, check=True)
    
    # Modify the .bim file to include updated marker IDs
    awk_command = (
        f"awk '{{print $1, $1\"\"$4\"\"$6\"_\"$5, $3, $4, $5, $6}}' OFS='\t' "
        f"Chr{i}_ma_filtered.bim.backup > Chr{i}_ma_filtered.bim"
    )
    subprocess.run(awk_command, shell=True, check=True)
```

```python
import subprocess
import os

# Define the directory where the .bim files are located
bim_dir = "plink"

for i in range(1, 23):
    # Construct file paths
    bim_file = f"{bim_dir}/Chr{i}_ma_filtered.bim"
    backup_file = f"{bim_file}.backup"
    
    # Check if the .bim file exists
    if not os.path.exists(bim_file):
        print(f"File not found: {bim_file}. Skipping chromosome {i}.")
        continue  # Skip to the next iteration
    
    # Backup the original .bim file
    backup_command = f"cp {bim_file} {backup_file}"
    subprocess.run(backup_command, shell=True, check=True)
    
    # Modify the .bim file to include updated marker IDs
    awk_command = (
        f"awk '{{print $1, $1\"\"$4\"\"$6\"_\"$5, $3, $4, $5, $6}}' OFS='\t' "
        f"{backup_file} > {bim_file}"
    )
    subprocess.run(awk_command, shell=True, check=True)

print("Processing completed.")
```

```python
import subprocess

for i in range(1, 23):
    # Step 1: Assign unique variant IDs
    set_ids_command = (
        f"plink2 --bfile plink/Chr{i}_ma_filtered "
        f"--set-all-var-ids '@:#' "
        f"--make-bed "
        f"--out plink/Chr{i}_unique_ids"
    )
    subprocess.run(set_ids_command, shell=True, check=True)

    # Step 2: Remove duplicate variants
    remove_dup_command = (
        f"plink2 --bfile plink/Chr{i}_unique_ids "
        f"--rm-dup exclude-all "
        f"--make-bed "
        f"--out plink/Chr{i}_unique_ids_nodup"
    )
    subprocess.run(remove_dup_command, shell=True, check=True)

    # Step 3: Perform LD pruning
    ld_prune_command = (
        f"plink2 --bfile plink/Chr{i}_unique_ids_nodup "
        f"--indep-pairwise 50 10 0.05 "
        f"--out plink/Chr{i}_ld_pruned"
    )
    subprocess.run(ld_prune_command, shell=True, check=True)

    # Step 4: Extract SNPs from LD pruning results
    filter_command = (
        f"plink2 --bfile plink/Chr{i}_unique_ids_nodup "
        f"--extract plink/Chr{i}_ld_pruned.prune.in "
        f"--make-bed "
        f"--out plink/Chr{i}_filtered_final"
    )
    subprocess.run(filter_command, shell=True, check=True)

print("Duplicate removal, LD pruning, and filtering completed successfully.")
```

```python
#Show individuals here.
import pandas as pd
import numpy as np

# Read the sample IDs, skipping the header line
sample_ids = pd.read_csv("GRM_Matrix.rel.id", delim_whitespace=True, skiprows=1, header=None, names=['FID', 'IID'])

# Read the GRM matrix
grm = pd.read_csv("GRM_Matrix.rel", delim_whitespace=True, header=None)

grm.columns = sample_ids['IID']
grm.index = sample_ids['IID']

# Read population information
pop_info = pd.read_csv(
    "20130606_sample_info.txt",
    sep='\t',
    usecols=['Sample', 'Population'],
    dtype={'Sample': str, 'Population': str}
)
pop_info.columns = ['IID', 'Population']

# List of unique populations
populations = pop_info['Population'].dropna().unique().tolist()

closest_pairs = {}

for pop in populations:
    # Get the samples belonging to this population
    pop_samples = pop_info[pop_info['Population'] == pop]['IID']
    pop_samples = [iid for iid in pop_samples if iid in grm.columns]
    
    if len(pop_samples) < 2:
        continue  # Skip populations with fewer than 2 samples
    
    # Subset the GRM matrix for this population
    grm_pop = grm.loc[pop_samples, pop_samples]
    
    # Set diagonal to NaN to ignore self-relatedness
    np.fill_diagonal(grm_pop.values, np.nan)
    
    # Assign unique names to the index and columns
    grm_pop.index.name = 'IID1'
    grm_pop.columns.name = 'IID2'
    
    # Unstack the matrix to get pairs
    grm_unstacked = grm_pop.unstack().dropna()
    grm_unstacked = grm_unstacked.reset_index()
    grm_unstacked.columns = ['IID1', 'IID2', 'Relatedness']
    
    # Remove duplicate pairs
    grm_unstacked['Pair'] = grm_unstacked.apply(lambda row: tuple(sorted([row['IID1'], row['IID2']])), axis=1)
    grm_unstacked = grm_unstacked.drop_duplicates('Pair')
    
    # Get the top 3 pairs with highest relatedness
    top_pairs = grm_unstacked.nlargest(3, 'Relatedness')[['IID1', 'IID2', 'Relatedness']]
    closest_pairs[pop] = top_pairs

# Display the results
for pop, pairs in closest_pairs.items():
    print(f"Population: {pop}")
    print(pairs.to_string(index=False))
    print("\n")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read PCA results
eigenval = pd.read_csv("PCA_Result.eigenval", delim_whitespace=True, header=None)
eigenvec = pd.read_csv("PCA_Result.eigenvec", delim_whitespace=True, header=None)

# Set column names for eigenvectors
num_pcs = eigenval.shape[0]
eigenvec.columns = ['FID', 'IID'] + [f'PC{i}' for i in range(1, num_pcs + 1)]

# Read population information
pop_info = pd.read_csv(
    "20130606_sample_info.txt",
    sep='\t',
    usecols=['Sample', 'Population'],
    dtype={'Sample': str, 'Population': str}
)
pop_info.columns = ['IID', 'Population']

# Population to Superpopulation mapping
population_to_superpop = {
    'ACB': 'AFR', 'ASW': 'AFR', 'ESN': 'AFR', 'GWD': 'AFR', 'LWK': 'AFR', 'MSL': 'AFR', 'YRI': 'AFR',
    'CLM': 'AMR', 'MXL': 'AMR', 'PEL': 'AMR', 'PUR': 'AMR',
    'CDX': 'EAS', 'CHB': 'EAS', 'CHS': 'EAS', 'JPT': 'EAS', 'KHV': 'EAS',
    'CEU': 'EUR', 'FIN': 'EUR', 'GBR': 'EUR', 'IBS': 'EUR', 'TSI': 'EUR',
    'BEB': 'SAS', 'GIH': 'SAS', 'ITU': 'SAS', 'PJL': 'SAS', 'STU': 'SAS'
}

# Add superpopulation information
pop_info['Superpop'] = pop_info['Population'].map(population_to_superpop)

# Merge PCA data with population information
pca_data = pd.merge(eigenvec, pop_info, on='IID')

# Convert PC columns to numeric type
pca_data['PC1'] = pd.to_numeric(pca_data['PC1'])
pca_data['PC2'] = pd.to_numeric(pca_data['PC2'])

# Plot 1: PCA by Continental Population
plt.figure(figsize=(12, 8))
superpop_colors = {
    'AFR': 'blue',
    'AMR': 'green',
    'EAS': 'red',
    'EUR': 'purple',
    'SAS': 'orange'
}

for superpop, color in superpop_colors.items():
    mask = pca_data['Superpop'] == superpop
    plt.scatter(
        pca_data.loc[mask, 'PC1'], 
        pca_data.loc[mask, 'PC2'],
        c=color,
        label=superpop,
        alpha=0.6,
        s=50
    )

plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.title('PCA of 1000 Genomes Data by Continental Population', fontsize=14)
plt.legend(title='Super Population', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.savefig('pca_continental.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
# Plot 2: PCA with Closest Individuals
# Collect IIDs of individuals to highlight
highlight_iids = set()
for pairs in closest_pairs.values():
    highlight_iids.update(pairs['IID1'])
    highlight_iids.update(pairs['IID2'])

pca_data['Highlight'] = pca_data['IID'].isin(highlight_iids)

plt.figure(figsize=(12, 8))

# Plot background points
mask_background = ~pca_data['Highlight']
plt.scatter(
    pca_data.loc[mask_background, 'PC1'],
    pca_data.loc[mask_background, 'PC2'],
    c='lightgrey',
    label='Others',
    alpha=0.5,
    s=50
)

# Plot highlighted points
mask_highlight = pca_data['Highlight']
plt.scatter(
    pca_data.loc[mask_highlight, 'PC1'],
    pca_data.loc[mask_highlight, 'PC2'],
    c='red',
    label='Top Closest Pairs',
    alpha=0.8,
    s=80
)

plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.title('PCA of 1000 Genomes Data with Closest Individuals Highlighted', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('pca_highlighted.png', dpi=300, bbox_inches='tight')
plt.show()

# Print summary statistics to verify data

print("\nPCA Summary Statistics:")
print(pca_data[['PC1', 'PC2']].describe())
```
